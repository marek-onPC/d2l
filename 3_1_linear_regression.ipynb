### Vectorization for Speed

```python
import torch
import time

n = 100000
a = torch.ones(n)
b = torch.ones(n)


c = torch.zeros(n)
t = time.time()

for i in range(n):
    c[i] = a[i] + b[i]

f'{time.time() - t:.5f} sec'
```

```python
t = time.time()
d = a + b
f'{time.time() - t:.5f} sec'
```

### The Normal Distribution and Squared Loss

```python
# Function to compute the nominal distribution (Gaussian)
import math
import numpy as np

def nominal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    result = p * np.exp(- (1 / (2 * sigma**2)) * (x - mu)**2 )

    return result
```

```python
from matplotlib import pyplot

# For visualization prepare inputs with NumPy
x = np.arange(-7, 7, 0.01)

# Mean and standard deviation pairs
params = [(0, 1), (0, 2), (3, 1)]

graph = pyplot.gca()

for mu, sigma in params:
    graph.plot(x, nominal(x, mu, sigma))

graph.legend(["mean 0, std 1", "mean 0, std 2", "mean 3, std 1"])
graph.grid()
```
